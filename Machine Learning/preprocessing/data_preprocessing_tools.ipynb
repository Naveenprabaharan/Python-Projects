# Data Preprocessing Tools

## Importing the libraries

```
#importing Library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
#Importing the dataset
dataset = pd.read_csv('Data.csv')
#assigning independet variable
x = dataset.iloc[:,:-1].values
#assigning dependetend variable
y = dataset.iloc[: , -1].values
```

```
print(x)
```

```
print(y)
```

## Taking care of missing data

```
#import SimpleImputer for insert a mean of all data to missing data 
from sklearn.impute import SimpleImputer
#calculate mean for missing data column
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
#assigning missing data
imputer.fit(x[ : , 1:3])
x[ : , 1:3] = imputer.transform(x[ : , 1:3])
```

```
print(x)
```

## Encoding categorical data

### Encoding the Independent Variable

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder


#decleare variable for encoding independent string into number
ct = ColumnTransformer(transformers=[('encode', OneHotEncoder(), [0])] ,remainder='passthrough')
#ecoding the "0"th ndex using OneHotEncoder methode ..remaining is 'no change'-passthrough

#fit the value into data
x = np.array(ct.fit_transform(x))

```

```
print(x)
```

### Encoding the Dependent Variable

```
#encoding string "yes""no" as 0 and 1
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

```
#importing train_test_split for spliting the data set into Trainging set and Test set
#Train set is used to train the model
#test set is used to test the model
from sklearn.model_selection import train_test_split

#x_train -- tain data set for X
#X_test - tets data set for X
#Y_train - tain data set for Y
#Y_test - tets data set for Y

X_train , X_test, Y_train , Y_test = train_test_split(x , y, test_size = 0.2 , random_state=1)

```

```
print(X_train)
```

```
print(X_test)
```

```
print(Y_train)
```

```
print(Y_test)
```

## Feature Scaling

```
#import StandardScaler for creatinf ferature scaling to avoid data misbehaviour
from sklearn.preprocessing import StandardScaler

#StandardScaler variable changed to sc
sc = StandardScaler()

#fit_transform => it is used to create sclaler value  and by using that sclaler value, tranform  our dataset value as per StandardScaler function value
X_train[ : ,3 : ] = sc.fit_transform(X_train[ : ,3 : ])

#transform => it is used existing sclaler value  and by using that sclaler value, tranform  our dataset value as per StandardScaler function value
#becase we have to use same scaler value for both train and test .if differ scaller value then our result will not be that much accurate
X_test[ : , 3: ] = sc.transform(X_test[ : ,3 : ])
```
